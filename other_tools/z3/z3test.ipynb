```python
import time  
from z3 import *
```

```python
# p1 (freire1)
start = time.time()
x,y,r,a1,a2 = Reals('x y r a1 a2')
xrange = And(x>=-10, x<=10, y>=-10,y<=10,r>=-10, r<=10)
arange = And(a1>=-2, a1<=2, a2>=-2, a2<=2)

inv = 2*x+a2*r**2-a1*r-y<=0
inv_ineq = And(x>=0,y>=0)

guard1 = -1<=0
inv_f1 = 2*(x-r)+a2*(r+1)**2-a1*(r+1)-y<=0

loop_cond = r-x<=0
pre = -1<=0
post = -1<=0

cons1 = ForAll([x,y,r],Implies(And(pre,x==y/2, r==0, xrange), inv))
cons2 = ForAll([x,y,r],Implies(And(inv, guard1, loop_cond, xrange), inv_f1))
cons3 = ForAll([x,y,r],Implies(And(inv, Not(loop_cond),xrange),post))
               
s = Solver()
s.add( cons1, cons2, cons3, arange)
print(s.check())
print(s.model())
end = time.time()
print("Consuming Time: %.8f s\n"%((end-start)))
```

```python
# unicycle
start = time.time()
x1,x2,a1,a2 = Reals('x1 x2 a1 a2')
xvars = [x1,x2]
xrange = And(x1>=-5, x1<=5, x2>=-5,x2<=5)
arange = And(a1>=0.95, a1<=1.05, a2>=-2, a2<=2)

loop_cond = -1<=0

pre = x1**2+(x2-1)**2-1<= 0
post = x1**2 + (x2-1)**2 - 4<=0

inv = x1**2 + a1*x2**2 - 2*a1*x2 +a2<=0
guard1 = -1<=0

w = 1.0178 + 1.8721 * x1 - 0.0253 * x2
x1_f = x1 + 0.01*(1-x2*w)
x2_f = x2 + 0.01*x1*w
inv_f1 = x1_f**2 + a1*x2_f**2 - 2*a1*x2_f +a2<=0

inv_ineq = And(x1>=0,x2>=0)


cons1 = ForAll(xvars,Implies(And(pre, xrange), inv))
cons2 = ForAll(xvars,Implies(And(inv, guard1, loop_cond, xrange), inv_f1))
cons3 = ForAll(xvars,Implies(And(inv, Not(loop_cond),xrange),post))
               
s = Solver()
s.add( cons1, cons2, cons3, arange)
print(s.check())
print(s.model())
end = time.time()
print("Consume Time: %.8f s\n"%((end-start)))
```
